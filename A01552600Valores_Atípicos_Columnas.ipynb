```python
#carga de un archivo csv sin √≠dices
df=pd.read_csv('Valores_sin_nulos.csv')
df.head()
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df1=df.copy()
```

```python
valores_nulos=df1.isnull().sum().sum()
valores_nulos
```

#1

```python
#realizamos el diagrama de caja o bigote
fig = plt.figure(figsize =(10,6))
plt.boxplot(df1['ventas_precios_constantes'])
plt.title('Outliers de ventas_precios_constantes')
plt.show() #dibujamos el diagrama
```

```python
#M√©todo aplicando desviaci√≥n est√°ndar. Encuentro los valores extremos
y=df1['ventas_precios_constantes']
Limite_superior= y.mean() + 3*y.std()
Limite_inferior= y.mean() - 3*y.std()
print('Limite superior permitido', Limite_superior)
print('Limite inferior permitido', Limite_inferior)
```

```python
#Encontramos outliers
outliers = df1[(y>Limite_superior) | (y<Limite_inferior)]
outliers
```

```python
#obtenemos datos limpios
data_clean_ventas_precios_constantes = df1[(y<=Limite_superior) & (y>=Limite_inferior)]
data_clean_ventas_precios_constantes
```
